```python
predictors_cal = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
target_cal = ['quality']
```

```python
predictors = data[predictors_cal]
target = data[target_cal]
```

```python
predictors
```

```python
target
```

```python
corr = predictors.corr()
corr
```

```python
sns.heatmap(corr)
plt.title('Matriz de Correlación')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(arbol, filled=True, feature_names=predictors.columns, class_names=[str(i) for i in range(1, 11)], fontsize=6)
plt.title("Árbol de Decisión")
plt.show() 
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```

```python
precision = precision_score(y_test, predicciones, average='weighted')
precision
```

```python
recall = recall_score(y_test, predicciones, average='weighted')
recall
```

```python
f1 = f1_score(y_test, predicciones, average='weighted')
f1
```
